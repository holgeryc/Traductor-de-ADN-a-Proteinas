```
dic_ADN_ARN = {
    'C':'G',
    'G':'C',
    'A':'U',
    'T':'A',
    '\n':''
}
dic_ARN_ADN = {
    'C':'G',
    'G':'C',
    'A':'T',
    'U':'A',
    '\n':''
}
dic_ARN_PRO = {
    'UUU':'F',
    'UUC':'F',
    'UUA':'L',
    'UUG':'L',
    'UCU':'S',
    'UCC':'S',
    'UCA':'S',
    'UCG':'S',
    'UAU':'Y',
    'UAC':'Y',
    'UAA':'E',
    'UAG':'E',
    'UGU':'C',
    'UGC':'C',
    'UGA':'E',
    'UGG':'W',
    'CUU':'L',
    'CUC':'L',
    'CUA':'L',
    'CUG':'L',
    'CCU':'P',
    'CCC':'P',
    'CCA':'P',
    'CCG':'P',
    'CAU':'H',
    'CAC':'H',
    'CAA':'Q',
    'CAG':'Q',
    'CGU':'R',
    'CGC':'R',
    'CGA':'R',
    'CGG':'R',
    'AUU':'I',
    'AUC':'I',
    'AUA':'I',
    'AUG':'M',
    'ACU':'T',
    'ACC':'T',
    'ACA':'T',
    'ACG':'T',
    'AAU':'N',
    'AAC':'N',
    'AAA':'K',
    'AAG':'K',
    'AGU':'S',
    'AGC':'S',
    'AGA':'R',
    'AGG':'R',
    'GUU':'V',
    'GUC':'V',
    'GUA':'V',
    'GUG':'V',
    'GCU':'A',
    'GCC':'A',
    'GCA':'A',
    'GCG':'A',
    'GAU':'D',
    'GAC':'D',
    'GAA':'E',
    'GAG':'E',
    'GGU':'G',
    'GGC':'G',
    'GGA':'G',
    'GGG':'G',
    '\n':''
}
#convierte linea de ADN a una linea ARN
def Linea_ADN_ARN(linea):
  resultado = ""
  x = len(linea)
  for k in range(0,x):
    resultado += str(dic_ADN_ARN[linea[k]])
  return resultado

#convierte un texto ADN a un texto ARN
def ADN_ARN(f):
  resultado = open('/content/fasta/adn-arn.fasta', 'w')
  for linea in f.readlines():
    if linea.startswith( ">" ):
      resultado.write(linea + "\n")
    else:
      linea_transcripcion = Linea_ADN_ARN(linea)
      resultado.write(linea_transcripcion + "\n")

#convierte una linea ARN a proteinas cada 3 caracteres
def Linea_ARN_PRO(linea,inicio,fin):
  resultado = ""
  resultado_completo = ""
  x = len(linea)
  for k in range(0,x-2,3):
    llave = str(linea[k])+str(linea[k+1])+str(linea[k+2])
    traduccion = str(dic_ARN_PRO[llave])
    if traduccion == "M":
      inicio = True
    elif traduccion == "E":
      if inicio:
        fin = True
    elif inicio and not fin:
      resultado+=str(traduccion)
    resultado_completo+=str(traduccion)
    print(traduccion, inicio, fin)
  return resultado,resultado_completo,inicio,fin

#convierte un texto ARN a dos textos de Proteinas, completa y con inicio-fin
def ARN_PROTEINA(f):
  resultado = open('/content/fasta/arn-proteina.fasta', 'w')
  resultado_completo = open('/content/fasta/arn-proteina-completo.fasta', 'w')
  inicio = False
  fin = False
  for linea in f.readlines():
    if linea.startswith( ">" ):
      resultado.write(linea + "\n")
      resultado_completo.write(linea + "\n")
    else:
      linea_traduccion,linea_traduccion_completo,inicio,fin = Linea_ARN_PRO(linea,inicio,fin)
      resultado.write(linea_traduccion + "\n")
      resultado_completo.write(linea_traduccion_completo + "\n")

adn = open('/content/fasta/adn.fasta', 'r')
ADN_ARN(adn)

adn_arn = open('/content/fasta/adn-arn.fasta', 'r')
ARN_PROTEINA(adn_arn)
```
